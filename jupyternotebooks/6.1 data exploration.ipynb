## import neccessary Libraries

```python
import os
import pandas as pd
import numpy as np
```

##Load data

```python
# load data from csv file
df = pd.read_csv("/content/drive/MyDrive/boat_data.csv")
print(df.shape)
print(df.info())
df
```

```python
#Check for missing data
missing_data_stats=df.isnull().sum()
print(missing_data_stats)
```

###Check for duplicate data

```python
duplicates_stats=df.duplicated().sum()
print(duplicates_stats)
```

##Perform data Wragling/Cleansing

```python
# Data Cleaning

# fill the Length an missing valus with median
df['Length'] = df['Length'].fillna(df['Length'].median())
#Removing rows with missing data
df=df.dropna()
print(df.isnull().sum())
```

###separte currency and price

```python
# split the price column
df['Currency'] = df.loc[:,'Price'].str.slice(0,3)
df.head(2)
```

### string Alphabetical values from the price column

```python
# split the price
df['Price'] = df['Price'].str.replace(r'[A-Z\D\W]', '')
```

###Adjust the data type

```python
# change data type of price to float
df['Price'] = df['Price'].astype('float')
```

###Convert the prices to a single currency

```python
###Convert the prices to a single currency
def eurprice(Price,Currency):
    if Currency=='CHF':
        return Price * 1.02
    elif Currency=='DKK':
        return Price * 0.13
    elif Currency=='Â£':
        return Price * 1.17
    else:
        return Price
df['EUR Price'] = df.apply(lambda x: eurprice(x.Price, x.Currency), axis=1)
df.head(2)
```

##Extract the country and city from the location

```python
df[['Country', 'City', 'Standing_point']] = df['Location'].str.split('Â»', n=2, expand=True)
df.head(2)
```

### removing redudant columns

```python
# extract Country and City from Location
# Drop the redundant columns
df = df.drop(['Price','Currency','Location','Standing_point'], axis = 1)
df.head(2)
```
